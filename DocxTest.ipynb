```python
import docx.shared
import re
from docx import Document
from docx.shared import Inches
from docx.shared import Pt
from docx.oxml.ns import qn
from docx.enum.style import WD_STYLE_TYPE
from docx.enum.text import WD_ALIGN_PARAGRAPH
from selenium import webdriver
from selenium.webdriver.common.by import By

def vulnerabilities_search_CVE(soft_name):
    driver = webdriver.Chrome()
    driver.get("https://cve.mitre.org/cve/search_cve_list.html")
    elem = driver.find_element(By.NAME, "keyword")
    elem.send_keys(soft_name)
    elem.submit()
    rows = len(driver.find_elements(By.XPATH, value = '//*[@id="TableWithRules"]/table/tbody/tr'))
    print ("rows = ", rows)
    cols = 2
    records = []
    if (rows == 0):
        records.append([])
        records[0].append("Отсутствует")
        records[0].append("Отсутствует")
        print (records)
        driver.close()
        return records
    else:
        for r in range(1, rows+1):
            records.append([])
            for c in range(1, cols+1):
                value = driver.find_element(By.XPATH, value = '//*[@id="TableWithRules"]/table/tbody/tr['+str(r)+']/td['+str(c)+']').text
                records[r-1].append(value)
        print(driver.title)
        #print ("records: ", records)
        driver.close()
        print (records)
        return records



def iter_heading(paragraphs):
    for paragraph in paragraphs:
        isItHeading=re.match('Heading ([1-9])',paragraph.style.name)
        if isItHeading:
            yield int(isItHeading.groups()[0]),paragraph

            
def addHeaderNumbering(document):
    hNums=[0,0,0,0,0]
    for index,hx in iter_heading(document.paragraphs):
        # ---put zeroes below---
        for i in range(index+1,5):
            hNums[i]=0
        # ---increment this---
        hNums[index]+=1
        # ---prepare the string---
        hStr=""
        for i in range(1,index+1):
            hStr+="%d."%hNums[i]
        # ---add the numbering---
        hx.text=hStr+" "+hx.text
        #Добавить здесь изменение шрифта, размера и цвета.

soft_name = input("Название ПО: ")
document = Document()

'''
styles = document.styles
styles['Heading 1'].font.name = 'Courier New'
styles['Heading 1'].font.size = Pt(26)
styles['Heading 1'].font.bold = True
styles['Heading 1'].font.color.rgb = docx.shared.RGBColor(0, 0, 0)
'''

'''
headOne = document.add_heading('', level=1)
run = headOne.add_run("Heading, level 1")
run.font.name = 'Arial'
run.font.size = Pt(16)
run.font.color.rgb = docx.shared.RGBColor(0, 0, 0)
'''
#head_one = document.add_paragraph("", "Heading 1")
#heade_one_text = head_one.add_run("Heading level 1")
#heade_one_text.font.name = 'Arial'
#heade_one_text.font.size = Pt(16)
#heade_one_text.font.color.rgb = docx.shared.RGBColor(0, 0, 0)


head_two = document.add_heading('', 2)
text_head_two = head_two.add_run(soft_name)
text_head_two.font.name = 'Arial'
text_head_two.font.size = Pt(12)
text_head_two.font.color.rgb = docx.shared.RGBColor(0, 0, 0)
text_head_two.font.all_caps = True

'''
userStyleHeadOne = document.styles.add_style('UserHead1', WD_STYLE_TYPE.PARAGRAPH)
userStyleHeadOne.font.name = 'Arial'
userStyleHeadOne.font.size = Pt(16)
userStyleHeadOne.font.underline = False
userStyleHeadOne.paragraph_format.alignment = WD_ALIGN_PARAGRAPH.LEFT
'''
'''
p = document.add_paragraph('A plain paragraph having some ')
p.add_run('bold').bold = True
p.add_run(' and some ')
p.add_run('italic.').italic = True
'''
#document.add_paragraph(headOne, style='UserHead1')
#document.add_paragraph('Intense quote', style='Intense Quote')

'''
document.add_paragraph(
    'first item in unordered list', style='List Bullet'
)
document.add_paragraph(
    'first item in ordered list', style='List Number'
)
'''
'''
records = (
    (3, '101', 'Spam'),
    (7, '422', 'Eggs'),
    (4, '631', 'Spam, spam, eggs, and spam')
)
'''

records = vulnerabilities_search_CVE(soft_name)
table = document.add_table(rows = 1, cols=2)
hdr_cells = table.rows[0].cells
hdr_cells[0].text = 'Идентификатор CVE'
hdr_cells[1].text = 'Описание уязвимостей'
#hdr_cells[2].text = 'Уровень опасности уязвимости'
for qty, desc in records:
    row_cells = table.add_row().cells
    row_cells[0].text = qty
    row_cells[1].text = desc
    #row_cells[2].text = desc

document.add_page_break()

#addHeaderNumbering(document) # проставляет нумерацию заголовкам, меняя шрифт, размер и цвет на дефолтный

document.save('demo.docx')
print ("OK")
```
